```python
def type_distribution(data):
    f_cnt = 0
    p_cnt = 0
    b_cnt = 0
    in_cnt = 0
    out_cnt = 0
    for d_str in data :
        if str(d_str) == 'None':
            pass
        else :
            f_cnt += d_str.count('음식')
            p_cnt += d_str.count('사람')
            b_cnt += d_str.count('음료')
            in_cnt += d_str.count('실내')
            out_cnt += d_str.count('실외')
             
    return f_cnt , p_cnt, b_cnt, in_cnt, out_cnt
```
